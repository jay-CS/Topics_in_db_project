```python
# print(len(articles_text),sum([len(arr) for arr in ALL_URLS]))

    
### SENTIMENT ANALYSIS
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer, VaderConstants
import csv 

lexicon_dict = {}

# with open('stock_market_lexicon.csv') as csv_file:
#     reader = csv.reader(f)
#     next(reader)  # Skip the header row
#     for row in reader:
#         word, pos, pos_score, neg_score = row
#         lexicon_dict[pos] = float(pos_score) - float(neg_score)

# Set up NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')


results = []
# Preprocess the tweets
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

#
for stock_articles in all_stock_article_content:
    print("NEW STOCK ANALYSIS ~~~~~~~~~~~~~")
    scores = []
    for article in stock_articles:
        # Tokenize tweet
        tokens = word_tokenize(article.lower())
        
        # Remove stop words
        tokens = [token for token in tokens if token not in stop_words]
        
        # Lemmatize tokens
        lematized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
        
        # Convert tokens back to string
        article_text = ' '.join(lematized_tokens)
        
        # Analyze sentiment using VADER
        sia = SentimentIntensityAnalyzer()
        sia.lexicon.update(lexicon_dict)
        sentiment = sia.polarity_scores(article_text)
        scores.append([sentiment['neg'],sentiment['pos']])
    results.append([sum([x[0] for x in scores])/len(scores), sum([x[1] for x in scores])/len(scores)])

print(results)
    
```
